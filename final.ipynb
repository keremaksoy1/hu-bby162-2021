```python
dosya = "/content/drive/MyDrive/Colab Notebooks/Final.txt"
print("Kütüphane Kataloglama Sistemine Hoşgeldiniz")

def KitapSırala():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()

def KitabiKaydet():
  kitapAdi = input("Kitap Adını Giriniz.")
  kitapYazari = input("Kitabin Yazarını Giriniz.")
  kitapBasımtarihi = input("Kitabin Basım Tarihini Giriniz.")
  kitapYayınevi = input("Kitabin Yayınevi Bilgisini Giriniz.")
  kitapIsbn = input("Kitabin ISBN Bilgilerini Giriniz.")
  f = open(dosya, 'a' )
  f.write(kitapAdi + " , " + kitapYazari + " , " + kitapBasımtarihi + " , " + kitapYayınevi + " , " + kitapIsbn + '\n')
  print("Kaydı Başarıyla Eklediniz")
  f.close()
  
islemSec = ["sıralama", "kaydet", "çıkış"]
while islemSec:
  girilen_islem = input("Katalogdaki kitap listesini görüntülemek için 'sıralama'\n"
      "Yeni kitap kaydı eklemek için 'kaydet' görevini yazınız.\n"     
      "Sistemden çıkmak için 'çıkış' görevini yazınız.\n"
      "Yapmak istediğiniz işlem için ilgili görevi yazınız: ")
  
  if girilen_islem == islemSec[0]:
    print("Kütüphane kataloğuna hoş geldiniz..")
    KitapSırala()
  elif girilen_islem == islemSec[1]:
    print("Kataloğa yeni bir kayıt giriniz.")
    KitabiKaydet()
  elif girilen_islem == islemSec[2]:
    break
```
